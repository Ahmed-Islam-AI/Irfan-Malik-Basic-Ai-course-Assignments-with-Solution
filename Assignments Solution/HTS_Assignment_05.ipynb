# **Alternative method for Image Classification**

```python
from transformers import ViTImageProcessor, ViTForImageClassification  # Importing libraries and processor from transformers for Image processing and classification
from PIL import Image                   # Importing 'Python Image library' to work with image
import requests                         # Importing requests library to send HTTp requests and handle responses

url = 'http://images.cocodataset.org/val2017/000000039769.jpg'         # Specify the URL of the image you want to classify
image = Image.open(requests.get(url, stream=True).raw)                 # Retrieves an image from URL and opens it, using the Image module from the PIL library.

processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')      # Create an instance of the ViTImageProcessor from the 'google/vit-base-patch16-224' pre-trained model
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')  # Create an instance of the ViTForImageClassification model from the 'google/vit-base-patch16-224' pre-trained model


inputs = processor(images=image, return_tensors="pt")      # Preprocess the image using the ViTImageProcessor and convert it into tensors that are processed by model efficiently

outputs = model(**inputs)       # Pass the preprocessed image tensors to the ViTForImageClassification model for prediction
logits = outputs.logits         # Get the predicted class logits from the model's output


predicted_class_idx = logits.argmax(-1).item()    # Determine the predicted class index by selecting the class with the highest probability

print("Predicted class:", model.config.id2label[predicted_class_idx])     # Print the predicted class label using the model's configuration
```

### **Important terms for better understanding of the above code**
- **Processor:** It is necessary for the transformation of the image to make it compatible with the model being used. It involves resizing, normalizing and other preprocessing steps. Its output is in the form of dictionary that is then used by model.
- **Model:** It perfroms the actual  lassification on the image. It takes the image from processor and produce logits.
- **Logits:** It is raw, unnormalized data/output of a model, before applying any activation function or probability interpretation. It is used to measure the model's confidence. Logits are actually vector numerical values, each representing a specific class/category. For example, for any cat image logits = [2.3 -> cat, -1.1 -> bird, 0.5 -> dog]. Now, to obtain final predictions, we usually apply a 'software function' that converts logits into probabilities like probabilites = [0.79 -> cat, 0.05 -> bird, 0.16 -> dog]

```python
import torch                                   # Import the torch library for tensor operations
from diffusers import StableDiffusionPipeline  # Import the StableDiffusionPipeline from the diffusers module
from transformers import CLIPTextConfig        # Import the CLIPTextConfig class from the transformers module

model_id = "CompVis/stable-diffusion-v1-4"     # Specify the ID of the pre-trained model to be used
device = "cpu"                                 # Set the device to CPU for inference, on which model will be loaded

text_config = CLIPTextConfig()    # Create an instance of CLIPTextConfig with default settings

# Initialize the StableDiffusionPipeline with the pre-trained model and custom configurations
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,                     # Name of the pre-trained model that we want to use
    torch_dtype=torch.float32,    # Set the data type for torch tensors to float32
    low_cpu_mem_usage=False,      # Disable low CPU memory usage mode, it is disabled becuase code was not working on low cpu usage
    text_config=text_config       # Pass the CLIPTextConfig instance to configure text processing, its like giving pipeline specific instructions on how to handle the text
)

pipe = pipe.to(device)            # Move the pipeline to the specified device (CPU in this case)

prompt = "a beautiful horse run on a track. A rider ride on a horse."  # Set the input text prompt
image = pipe(prompt).images[0]    # Process the prompt using the pipeline and get the generated image

image.save("horse.png")  # Save the generated image to a file
image.show()                      # Display the generated image
```
