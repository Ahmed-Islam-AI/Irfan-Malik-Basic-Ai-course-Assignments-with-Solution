```python
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
data=pd.read_csv("/content/housing.csv")
```

# missing data

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
data.count()
```

# removing Unnecassory Columns Permanently

```python
features=data.drop(["longitude","latitude"], axis=1, inplace=True)
```

```python
objcat=["ocean_proximity"]

for colname in objcat:
  data[colname]=data[colname].astype('category')
```

# Model Ready

```python
data.shape
```

```python
data.select_dtypes(['category']).columns
```

```python
ocean_proximity=pd.get_dummies(data['ocean_proximity'], drop_first=True)
```

```python
data.head()
```

```python
data=pd.concat([data,ocean_proximity], axis=1)
data.head(5)
```

```python
data.drop('ocean_proximity', axis=1, inplace= True)
```

# **Objective:02 Machine Learning**

**Split the data into training and test sets**

```python
X = data.drop('median_house_value', axis=1)

y= data['median_house_value']   #y is the column we are trying to pridict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=101)
```

### Fit a logistic regression model using scikit-learn

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# predict
y_pred_lreg = model.predict(X_test)
print(y_pred_lreg)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print('Classification Model')
# Accuracy
print('--'*40)
logreg_accuracy = round(accuracy_score(y_test, y_pred_lreg) * 100,2)
print('Accuracy', logreg_accuracy,'%')
```
